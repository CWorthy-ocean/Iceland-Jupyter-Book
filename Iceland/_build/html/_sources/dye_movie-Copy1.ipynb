```python
import xarray as xr
import matplotlib.pyplot as plt
import xrscipy
import numpy as np
import cartopy.crs as ccrs
from xmovie import Movie

from celluloid import Camera 
import xarray as xr
import matplotlib.pyplot as plt
import xrscipy
import numpy as np
import cartopy.crs as ccrs
import xesmf as xe
from matplotlib import colorbar, colors
from matplotlib.cm import get_cmap
from roms_regrid import *
from density import *
```

```python
x=xr.open_mfdataset('/expanse/lustre/projects/cos100/uheede/roms/run/Iceland4_multidye/z_Iceland4_mulitdye_rnd.20120601130000.nc', combine='nested', concat_dim=["time"])
grid=xr.open_mfdataset('/expanse/lustre/projects/cos100/uheede/Iceland4_grd.nc')
mask_rho=grid['mask_rho']
dye1=x['dye1']
dye2=x['dye2']
dye3=x['dye3']
dye4=x['dye4']
```

```python
dye1_plot=dye1.isel(depth=0).load()
dye2_plot=dye2.isel(depth=0).load()
dye3_plot=dye3.isel(depth=0).load()
;
```

```python
from matplotlib import pyplot as plt
from celluloid import Camera
from IPython.display import HTML
fig = plt.figure()
levels=[0.1,10,0.1]
cmap1='Reds'
cmap2='Blues'
cmap3='Greens'
cmap4='Oranges'
k1=0
k2=1
k3=0.1
levels = np.arange(k1, k2, k3)
cmap = 'jet'
camera = Camera(fig)
for i in range(48):
    plt.contourf(dye1_plot.isel(time=i).where(dye1_plot.isel(time=i)!=0),levels, extend="both", cmap=cmap1)
    plt.contourf(dye2_plot.isel(time=i).where(dye2_plot.isel(time=i)!=0),levels, extend="both", cmap=cmap2)
    plt.contourf(dye3_plot.isel(time=i).where(dye3_plot.isel(time=i)!=0),levels, extend="both", cmap=cmap3)
 
    camera.snap()
    plt.contourf(mask_rho.where(mask_rho!=1),cmap='Greys')
    plt.title('2 dye lines over 48 hours')
animation = camera.animate()
HTML(animation.to_html5_video())
```
